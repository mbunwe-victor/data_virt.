```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 400)
```

```python
def g(x):
    # the basic quadratic function and a plot of graph x and y
    return x**2+2*x+5

plt.plot(x, g(x))
plt.title("Plot of g(x) = x^2 + 2x + 5")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid(True)
plt.show()
```

```python
def h(x):
    ## running a sine function and plotting its graph
    return np.sin(x)

plt.plot(x, h(x))
plt.title("Plot of h(x) = sin(x)")
plt.xlabel("x")
plt.ylabel("h(x)")
plt.grid(True)
plt.show()
```

```python
def i(x):
    # Doing  a combination of the Quadratic and Sine function
    return np.sin(x**2 + 2*x + 5)

plt.plot(x, i(x))
plt.title("Plot of i(x) = sin(x^2 + 2x + 5)")
plt.xlabel("x")
plt.ylabel("i(x)")
plt.grid(True)
plt.show()
```

```python
def j(x):
    ## running an exeponential decay and presenting it on the graph
    return np.exp(-0.1*x)

plt.plot(x, j(x))
plt.title("Plot of j(x) = e^{-0.1x}")
plt.xlabel("x")
plt.ylabel("j(x)")
plt.grid(True)
plt.show()
```

```python
def f(x):
    ## here we want to do a combination of all the functions 
    ## and representing it on a graph
    return (np.sin(x**2 + 2*x + 5))*(np.exp(-0.1*x))

plt.plot(x, f(x))
plt.title("Plot of f(x) = sin(x^2 + 2x + 5) x e^{-0.1x}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
